```python
# Specify the URL of the YouTube video
import cgi 

form = cgi.FieldStorage()
youtube_url = form.getvalue("video-link")

# Extract the video ID from the URL
video_id = youtube_url.split('=')[1]

# Download the video file
yt = pytube.YouTube(youtube_url)
video = yt.streams.filter(progressive=True, file_extension='mp4').first()
video_filename = f'video1.mp4'
video.download(output_path='video', filename=video_filename)
```

```python
# Open the downloaded video file
video_path = f'video/{video_filename}'
video = cv2.VideoCapture(video_path)

# Create a folder named "frames" in the "video" folder if it doesn't exist
if not os.path.exists(f'video/frames'):
    os.makedirs(f'video/frames')
```

```python
# Set frame interval to 2 seconds
frame_interval = 2

# Initialize frame count and current time
count = 0
current_time = 0

# Loop through video file
while video.isOpened():

    # Set the frame pointer to the frame at the current time
    video.set(cv2.CAP_PROP_POS_MSEC, current_time)

    # Read a single frame from the video
    ret, frame = video.read()

    # Check if the frame was successfully read
    if not ret:
        break

    # Save the frame as an image file in the "frames" folder
    cv2.imwrite(f'video/frames/frame{count}.jpg', frame)

    # Increment frame count and current time by the frame interval
    count += 1
    current_time += frame_interval * 1000
```

```python
# Release the video file
video.release()
```
